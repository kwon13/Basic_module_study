## 1. 선 그래프 그리기
선 그래프는 시간에 따라, 숫자에 따라 변하는 관계를 시각화 하는데 유용합니다

plt.plot(x,y,fmt): x의 값, y의 값을 인자로 넣어주면 선 그래프를 그려줍니다 (fmt는 그래프를 꾸미는데 사용됩니다)

```python
data1=[10,20,30,50,37]
plt.plot(data1) # x의 값을 입력하지 않으면 자동으로 0부터 y개수에 맞게 생성됩니다
```

```python
# [<matplotlib.lines.Line2D at 0x7fbd085477f0>] <-그래프의 객체 정보 없이 출력하기 이해서는 plt.show()를 추가합니다
import numpy as np
random_y=np.random.randn(10)
plt.plot(random_y)
plt.show()
```

```python
x_data=np.arange(-10,11)
func=x_data**2+10
plt.plot(x_data, func)
plt.show()
```

```python
# 하나의 창에 여러 그래프 그리기 
x_data=np.arange(-4,5)
y1=x_data*3
y2=x_data**2
y3=5*x_data+10
plt.plot(x_data,y1, x_data,y2, x_data,y3)
plt.show()
```

```python
# plt.figure(figsize(m,n)): 새로운 그래프 창을 생성합니다 figsize=그래프의 크기를 지정합니다
x_data=np.arange(50)
y1=x_data**3
plt.figure(figsize=(5,4))
plt.plot(x_data,y1)
y2=x_data*-1
plt.figure(figsize=(3,3))
plt.plot(x_data,y2)
plt.show()
```

```python
# plt.subplot(m,n,p): 하나의 창에 구역을 나눠 여러 그래프를 그려줍니다 
# m x n 행렬로 이뤄진 그래프 중에서 p번째에 그려지게 됩니다

x=np.arange(0,10,0.1)
y1=x**3+10
y2=x*5
y3=np.sin(x)
y4=np.exp(x)

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.show()
```

```python
# plt.xlim(xmin,xmax), plt.ylim(ymin, ymax): 그래프의 출력범위 지정하기(특정 범위 확대하기)
x=np.arange(-4,4,0.5)
y1=x**3+3
y2=5*x

plt.plot(x,y1, x,y2)
plt.xlim(0,3)
plt.ylim(2,13)
```

## 2. 그래프 꾸미기
이전 plot에서 3번째 인자 fmt는 그래프를 꾸미는데 사용된다고 했습니다  
fmt=color, line_style, marker
3가지의 옵션이 있습니다

|컬러 약어|컬러|
|:---:|---|
|b|<span style="color:blue">파란색</span>|
|g|<span style="color:green">녹색</span>|
|r|<span style="color:red">빨간색</span>|
|c|<span style="color:cyan">청녹색</span>|
|m|<span style="color:magenta">자홍색</span>|
|y|<span style="color:yellow">노란색</span>|
|w|<span style="color:white">흰색</span>|
|k|<span style="color:black">검은색</span>|

|선 스타일 약어|선 스타일|
|:---:|---|
|:|점선(………)|
|-|실선(______)|
|--|파선(-----)|

|마커 약어|마커|
|:---:|---|
|^|삼각형|
|o|원|
|s|사각형|
|*|별|
|+|더하기|
|.|점|

```python
x=np.arange(1,6)
y1=x/2+1
y2=x/2+3
y3=x/2+5
plt.plot(x,y1,'-y', x,y2,':m', x,y3,'>r')
plt.show()
```

```python
x=np.arange(0,10,0.5)
y=x**4
plt.plot(x,y,'o--k')
plt.show()
```

이외에도 그래프의 라벨, 제목, 범례, 문자열, 격자 표시를 할 수 있습니다

```python
# plt.xlabel(), plt.ylabel(): 그래프의 라벨을 표시합니다
x=np.arange(0,10,0.5)
y=2*x**3
plt.plot(x,y,'-m')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

만약 한글을 입력했고 x의 범위가 -라면 아래와 같이 설정해 주세요

```python
# 맥 기준 한글폰트 설정
from matplotlib import rc
rc('font', family='AppleGothic')
#rc('font', family='NanumGothic') 윔도우용
plt.rcParams['axes.unicode_minus'] = False # - 부호 표시

x=np.arange(-4,10,0.5)
y=2*x**3

plt.plot(x,y,'-m')
plt.xlabel('x축')
plt.ylabel('y축')
```

```python
#  plt.title(): 그래프의 제목을 표시합니다
# plt.text(x,y,'~'): x,y에 텍스트를 추가합니다
plt.title('3차함수')
plt.plot(x,y,'g')
plt.xlabel('x-axis')
plt.ylabel('y-aixs')
plt.text(0,1000,'3차함수')
plt.grid() # 격자 표시
```

```python
# plt.legned([]): 그래프의 범례를 추가합니다
x=np.arange(0,10,0.1)
y1=x**2
y2=x*5
y3=x

plt.plot(x,y1,'r', x,y2,'g', x,y3,'b')
plt.legend(['2제곱','5X','X'])
plt.xlim(0,6)
plt.ylim(0,40)
plt.show()
```

## 3. 산점도
산점도 그래프는 두 개의 요소로 이루어진 집합과의 관계를 시각화 하는데 유용합니다  

plt.scatter(x,y,fmt): x의 값, y의 값을 인자로 넣어주면 산점도 그래프를 그려줍니다 

```python
height=[165,177,160,180,185,155,172]
weight=[62,67,55,74,90,43,64]

plt.scatter(height, weight)
plt.xlabel('키')
plt.ylabel('몸무게')
plt.grid()
plt.show()
```

```python
# 사이즈와 새깔을 지정합니다
size=np.random.randint(1,8,(7))*100
colors=['r','g','b','c','m','k','y']
plt.scatter(height,weight,s=size,c=colors)
plt.show()
```

```python
city=['서울','인천','대전','대구','을신','부신','광주']

#위도와 경도
lat=[37.65,37.45,36.35,35.87,35.53,35.18,35.16]
lon=[126.97,126.70,127.38,128.60,129.31,129.07,126.85]

#인구 밀도
pop_den=[16154,2751,2839,2790,1099,4454,2995]

size=np.array(pop_den)*0.2
colors=['r','g','b','c','m','k','y']

plt.scatter(lon,lat,s=size,c=colors)

for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name)
plt.show()
```

## 4. 막대 그래프
항목의 수량이 많고 적음을 시각화 하는데 유용합니다

plt.bar(x,y,fmt): x의 값, y의 값을 인자로 넣어주면 막대 그래프를 그려줍니다

```python
plt.title('몸무게')
member_id=['p1','p2','p3','p4']
before_w=[65,77,60,80]

colors=['m','g','c','y']

index=np.arange(len(member_id))
plt.bar(index,before_w,color=colors, tick_label=member_id, width=0.6)
plt.show()
```

```python
#기로 믹대 그래프를 그립니다( width불가)
plt.barh(index,before_w,color=colors, tick_label=member_id)
```

```python
# 두 데이터를 하나의 창에서 비교합니다
after_w=[62,67,55,74]

barWidth=0.4
plt.bar(index,before_w, align='edge', width=barWidth, label='before')
plt.bar(index+barWidth,after_w,color='c', align='edge', width=barWidth, label='after')
plt.xticks(index+barWidth,member_id)
plt.legend()
plt.show()
```

## 5. 히스토그램
데이터가 어떤 분포를 갖는지 볼 때 이용합니다

plt.hist(x,bins=class): 변량 x의 값을 넣어주면 히스토그램을 그려줍니다(class를 입력하지 않으면 자동으로 x개수만큼 설정됩니다)

```python
math=np.random.randint(80,100,(10))
plt.hist(math, bins=8)
plt.grid()
plt.show()
```

## 6. 파이 그래프
데이터에서 각 항목이 차지한 비율을 비교할 때 사용합니다

plt.pie(x): 차지하는 비율 x의 값을 넣어주면 파이 그래프를 그려줍니다

```python
#그래프에서 차지하고 있는 비율
p = [24, 26, 10, 40]
# 영역 이름
n = ['Apple', 'Samsung', 'LG', 'ASUS']
# autopct는 표시할 숫자의 형식입니다.
plt.pie(p, labels=n, autopct='%.1f%%')
plt.show()
```

```python
# Asus만 강조합니다
explode=[0,0,0,0.15]
plt.pie(p, labels=n, autopct='%.1f%%',startangle=90,explode=explode, shadow=True)
plt.figure(figsize=(5,5))
```
