```python
# pandas를 이용해 표를 읽고 가공하는 방법을 배웠습니다. 
# Matplotlib은 우리에게 데이터를 쉽게 볼 수 있도록 표, 그래프 등 다양한 방법으로 표현합니다. 
# Jupyter Notebook을 사용하면 훨씬 편리하게 볼 수 있습니다! 
# 실행하기 전 어떤 결과가 나올지 생각해 보세요!

# 우선 matplotlib.pyplot 모듈을 불러옵니다, matplotlib.pyplot는 너무 길기 때문에 plt로 축약해서 표현합니다.
import matplotlib.pyplot as plt

# 만약 모듈이 없다고 뜨면 터미널 PS C:\Users\파일이름> pip install matplotlib를 쳐주세요
```

```python
# plt.plot(): 괄호 안 인자에 값을 리스트 형식으로 적고, plt.show()라고 치면 그래프를 그려줍니다.
# 인자의 리스트 값이 [a,b],[c,d]라면 (a,c),(b,d)의 점을 찍고 그래프를 그려줍니다.
%matplotlib inline 
# ^-- Cell Magic이라고 부르며 주피터노트북에서 그래프를 바로 보여주는 코드입니다. (주피터 보트북이 아닌 분들은 필요하지 않아요!)
plt.plot([1,2,3,4],[5,6,8,10]) # plt.plot([5,6,8...]) 이 형식은 x의 값이 [0,1,2...]으로 지정됩니다.
plt.show()
```

```python
# plt.plot() 함수는 직선뿐만이 아니라 다른 모양도 표현이 가능합니다.

#  방금 적은 plt.plot()인자로 'ro'(r:red, o:점을 의미)를 추가합니다. 
# 'g-'(g:green, -:선을 의미), 'b^'(b:blue, ^:삼각형), 'r--'(--:점선), 'rs'(s:사각형) 말고도 다양한 표현이 있습니다!
plt.plot([1,2,3,4],[5,6,8,10],'ro')

# plt.axis(): 각 축의 범위를 지정합니다. 괄호 안 [xmin, xmax, ymin, ymax]를 적어 지정합니다.
plt.axis([0, 10, 3, 20])
plt.show()
```

```python
# 예제로 여러 함수를 그려보겠습니다.
import numpy as np
x=np.arange(0,6)
plt.plot(x, 'g.', x**2,'r--', x**3, 'b')
plt.show()
# g:X그래프,r:X제곱 그래프,b:X3제곱 그래프 ->이해가 되지 않으시면 numpy 자료를 참고하세요! 
```

```python
# pyplot는 제목과 각 축의 이름, 범례를 붙일 수 있습니다. 
# plt.title(제목), plt.xlabel(축이름), plt.ylabel(축이름), plt.legend(범례) 함수를 사용하면 됩니다.
x=np.arange(0,5)
plt.plot(x, 'k:', x**2,'y--', x**3, 'c-.')
plt.xlabel('밑')
plt.ylabel('값') 
plt.title('1차,2차,3차 그래프')
plt.legend(['X함수', 'X^2함수', 'X^3함수'])
plt.show()

# 축을 한글로 설정하신다면 사각형 모양으로 깨져서 나옵니다.
# 해결하기 위해서는 한글을 지원하는 폰트를 다운받고 plt.rc('font', family='한글지원폰트이름')을 칩니다.
plt.rc('font', family='NanumGothicCoding')

# 저는 나눔고딕코딩폰트를 설치했습니다!
# 링크:https://github.com/naver/nanumfont (<--나눔고딕코딩폰트 다운받는 곳) 
# 폰트 설치 방법: https://tttsss77.tistory.com/209

# 추가로 vscode를 사용하신다면 편집기 폰트를 설정할 수 있습니다.
# 링크:https://tttsss77.tistory.com/97 (vscode 폰트 설정 방법)
```

```python
# 추가적으로 -부호가 깨지는 현상이 나타난다면 plt.rcParams['axes.unicode_minus'] = False 이 코드를 추가해 주세요 
x=np.arange(-10,11)
plt.plot(x,x**2)
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
plt.rcParams['axes.unicode_minus'] = False
```

```python
# plt.grid():그리드(격자)를 표시할 수 있습니다. 
# color= 색(#~: Hex code)
x = np.arange(0, 5)
plt.plot(x,'b-')
plt.plot(x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x**3, color='springgreen', marker='^', markersize=9)
plt.grid(True)
plt.show()
```

```python
# matplotlib은 평면 그래프 뿐만 아니라 다양한 그래프도 그릴 수 있습니다!

# plt.bar(): 막대 그래프를 그립니다. width= 막대의 넓이(1~0)
plt.bar([1, 2, 3], [10, 4, 8], width=0.3, color='#e35f62') 
plt.show()
```

```python
# plt.pie(): 원형 그래프를 그립니다.

#그래프에서 차지하고 있는 비율
p = [24, 26, 10, 40]
# 영역 이름
n = ['Apple', 'Samsung', 'LG', 'ASUS']
# autopct는 표시할 숫자의 형식입니다.
plt.pie(p, labels=n, autopct='%.1f%%')
plt.show()
```
